```python
# Inicjalizujemy pustą ramkę danych z odpowiednimi kolumnami
cd = pd.DataFrame(columns=cc.columns)

# Przypisujemy kolumny z dd do odpowiednich kolumn w cd
cd[['Nr księgi ucznia', 'Imiona', 'Nazwisko', 'Data urodzenia', 'Miejsce urodzenia', 'Pesel', 'e-mail ucznia']] = dd[['nr_ksiegi', 'Imię', 'Nazwisko', 'Data urodzenia', 'Miejsce ur', 'PESEL', 'e-mail']]

# Zostawiamy pozostałe kolumny puste, ponieważ nie mamy dla nich danych w dd
```

```python
cd.head()
```

```python
# Usuwamy '.0' na końcu wartości, konwertujemy na string
cd['Pesel'] = cd['Pesel'].astype(str).str.split('.').str[0]

# Uzupełniamy zera na początku do 11 cyfr
cd['Pesel'] = cd['Pesel'].str.zfill(11)

# Usuwamy '.0' na końcu wartości, konwertujemy na string
cd['Nr księgi ucznia'] = cd['Nr księgi ucznia'].astype(str).str.split('.').str[0]
```

```python
cd
```

```python
cd['Płęć'] = cd['Imiona'].apply(lambda x: 'k' if str(x).strip().endswith('a') else 'm')
```

```python
cd.fillna('', inplace = True)
```

```python
cd['Imiona'] = cd['Imiona'].str.title()
cd['Nazwisko'] = cd['Nazwisko'].str.title()
```
