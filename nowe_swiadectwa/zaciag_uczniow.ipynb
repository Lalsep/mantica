```python
from fuzzywuzzy import fuzz

def match_names(name, list_names, min_score=0):
    # Start with an empty list of matches
    matches = []

    # For each name in the list, calculate fuzzy matching score
    for name2 in list_names:
        # FuzzyWuzzy returns a ratio given two strings. The closer the ratio is to 100, the smaller the edit distance between the two strings
        score = fuzz.ratio(str(name), str(name2))  # Ensure the names are strings

        # If the score is higher than the minimal score append the match to the matches list
        if (score >= min_score):
            matches.append((name2, score))

    # Sort the matches by score in descending order
    matches = sorted(matches, key=lambda x: x[1], reverse=True)

    # If there are no matches return None
    if len(matches) == 0:
        return None

    # Otherwise return the name and score of the best match
    else:
        return matches[0]

list_names = cd.index.tolist()

# Use the function
df['Match'] = df.index.map(lambda x: match_names(str(x), list_names, min_score=80))  # Ensure the index is a string
```
