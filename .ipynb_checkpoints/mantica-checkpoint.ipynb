```python
df = pd.read_clipboard()
df.drop(columns='L.p.', inplace = True)
```

```python
for e in df.columns:
    df[e] = df[e].str.strip().str.replace(' ', ' ')
df.dropna(inplace = True)
for e in ['Nazwisko ucznia', 'Imię/Imiona']:
    df[e] = df[e].str.title()
df
```

```python
db = pd.read_csv('import_uczniow.csv', encoding='utf-8', index_col=0, sep=';')
db.columns
```

```python
df.columns
```

```python
# Mapowanie kolumn z 'db' na 'df'
column_mapping = {
    'Nazwisko ucznia': 'Nazwisko',
    'Imię/Imiona': 'Imiona',
    'e-mail': 'e-mail ucznia'
}

# Tworzenie nowej ramki danych 'dc' z tymi samymi kolumnami co 'db'
dc = pd.DataFrame(columns=db.columns)

# Przenoszenie danych z 'df' do 'dc' zgodnie z mapowaniem kolumn
for df_column, db_column in column_mapping.items():
    if df_column in df.columns and db_column in dc.columns:
        dc[db_column] = df[df_column]
```

```python
dc.fillna('', inplace = True)
```

```python
dc.to_csv('import_mantica.csv', sep =';', index=False, encoding='utf-8')
```

```python
dx = pd.read_clipboard()
```

```python
# Czyszczenie danych w kolumnach, które są używane do połączenia
for column in ['Nazwisko ucznia', 'Imię ucznia']:
    if column in dx.columns:
        dx[column] = dx[column].str.strip().str.title()
    if column in dc.columns:
        dc[column] = dc[column].str.strip().str.title()

# Jeśli kolumna 'Drugie imię ucznia' istnieje, zastąp NaN pustymi ciągami znaków
if 'Drugie imię ucznia' in dx.columns:
    dx['Drugie imię ucznia'] = dx['Drugie imię ucznia'].fillna('')

# Łączenie 'Imię ucznia' i 'Drugie imię ucznia' w jednej kolumnie
if 'Drugie imię ucznia' in dx.columns:
    dx['Imię ucznia'] = dx[['Imię ucznia', 'Drugie imię ucznia']].apply(' '.join, axis=1)

# Przygotowujemy ramkę danych 'dx' do połączenia, usuwając niepotrzebne kolumny
dx_subset = dx[['Nazwisko ucznia', 'Imię ucznia', 'Data urodzenia', 'Miejsce urodzenia']]

# Połączenie dwóch ramek danych
dc = pd.merge(dc, dx_subset, on=['Nazwisko ucznia', 'Imię ucznia'], how='left')

# Zastąpienie NaN wartościami domyślnymi
dc['Data urodzenia'] = dc['Data urodzenia'].fillna('Nieznane')
dc['Miejsce urodzenia'] = dc['Miejsce urodzenia'].fillna('Nieznane')
```
