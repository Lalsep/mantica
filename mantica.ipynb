```python
df = pd.read_excel('Lista uczniów SP_LO_2022_2023 propozycje ocen rocznych.xlsx', usecols='A:E', skiprows=1, index_col=0, sheet_name = '1 SP uzup')
df.dropna(subset = ['L.p.'], inplace = True)
df.drop(columns='L.p.', inplace = True)
df.reset_index(inplace = True, drop = True)
df
```

```python
for e in df.columns:
    df[e] = df[e].str.strip().str.replace(' ', ' ')
df.dropna(inplace = True)
for e in ['Nazwisko ucznia', 'Imię/Imiona']:
    df[e] = df[e].str.title()
df['e-mail'] = df['e-mail'].str.lower()
df
```

```python
db = pd.read_excel('Uczniowie KNO 2022-2023 - aktualna.xlsx', sheet_name='1 sp uzup', index_col=0, usecols='A:F')
lista = list()
for e in db.columns:
    e = e.strip().replace('..','')
    lista.append(e)
db.columns=lista
```

```python
db['Data urodzenia'] = pd.to_datetime(db['Data urodzenia'])
```

```python
d = pd.read_csv('import_uczniow.csv', sep = ';', encoding='windows-1250')
d.columns
```

```python
# Mapowanie kolumn z 'db' na 'df'
column_mapping = {
    'Nazwisko ucznia': 'Nazwisko',
    'Imię/Imiona': 'Imiona',
    'e-mail': 'e-mail ucznia'
}

# Tworzenie nowej ramki danych 'dc' z tymi samymi kolumnami co 'db'
dc = pd.DataFrame(columns=db.columns)

# Przenoszenie danych z 'df' do 'dc' zgodnie z mapowaniem kolumn
for df_column, db_column in column_mapping.items():
    if df_column in df.columns and db_column in dc.columns:
        dc[db_column] = df[df_column]
```

```python
dc.fillna('', inplace = True)
```

```python
dc
```

```python
# Zmieniamy nazwy kolumn w db i df na te same, co w d
db = db.rename(columns={
    'Nazwisko ucznia': 'Nazwisko',
    'Imię ucznia': 'Imiona',
    'Drugie imię ucznia': 'Imiona',
    'Data urodzenia': 'Data urodzenia',
    'Miejsce urodzen': 'Miejsce urodzenia'
})

df = df.rename(columns={
    'Nazwisko ucznia': 'Nazwisko',
    'Imię/Imiona': 'Imiona',
    'e-mail': 'e-mail ucznia'
})

mantica = db.merge(df, left_on=['Nazwisko ucznia', 'Imię ucznia'], right_on=['Nazwisko ucznia', 'Imię/Imiona'], how='outer')


# Dopisujemy brakujące kolumny z d
for column in d.columns:
    if column not in mantica.columns:
        mantica[column] = np.nan
```

```python
mantica.reset_index(inplace = True)
mantica
```

```python
mantica.to_csv('mantica_import.csv', sep = ';', encoding='utf-8', index=False)
```
